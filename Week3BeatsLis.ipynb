```python
%matplotlib inline

from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,2*np.pi,0.01)
A = 1
B = 1
omega1 = np.pi
```

```python
def pltbeats(omega1, omega2):
    fig = plt.figure(figsize=(15, 8))
    plt.plot(t,A*np.sin(t*omega1)+(A*np.sin(t*omega2)))
    plt.ylim(-2.1,2.10)
    plt.xlim(0, 2*np.pi)
    plt.xlabel("time, s")
    plt.ylabel("Amplitude, arb")
```

```python
interact(pltbeats, omega1 =(40,50,0.1), omega2 = (40,50,0.1));
```

The high frequency component is: 
  
  
$$\sin{\left(\frac{(\omega_{1}+\omega_{2})t}{2}\right)}$$

The low frequency component (amplitude) is:  
  
  
$$\cos{\left(\frac{(\omega_{1}-\omega_{2})t}{2}\right)}$$

The beat of the angular frequency is given by:  
  
$$\omega_{beat} = \vert\omega_{1}-\omega_{2}\vert$$

```python
def pltlis(omegax, omegay, phix, phiy):
    fig = plt.figure(figsize=(8, 8))
    x = A*np.sin(t*omegax+phix)
    y = A*np.sin(t*omegay+phiy)
    plt.plot(x,y)
    plt.ylim(-1.1,1.10)
    plt.xlim(-1.1,1.1)
    plt.xlabel("x-displacement")
    plt.ylabel("y-displacement")
```

## Lissajous figures  
When two SHMs at right angles are combined, the path traced out will be a Lissajous figure.  


* The Lissajous figure can be described using two parametric equations  
  
$$x = \sin({\omega_{x} t} + \phi_{x})\quad \textrm{and} \quad y = \sin({\omega_{y} t + \phi_{y}})$$

* As $x$ and $y$ are orthogonal to each other, they move independently in an SHM motion.

```python
interact(pltlis, omegax =(1,10,1), omegay = (1,10,1), phix = (0,np.pi,np.pi/4),  phiy = (0,np.pi,np.pi/4));
```
